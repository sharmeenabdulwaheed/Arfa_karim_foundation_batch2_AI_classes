```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("E:\\iris.csv")
```

```python
df.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=10)
```

```python
features=df[["sepal length","sepal width",	"petal length",	"petal width"]]
```

```python
labels=df["specie"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test =train_test_split(features, labels, test_size=0.3, random_state=4)
```

```python
#train_data: X_train, y_train
#test_data: X_test, y_test
```

```python
knn.fit(X_train, y_train)
```

```python
knn.predict([[4.6,7,1.4,5.6]])
```

```python
X_test
```

```python
predictions=knn.predict(X_test)
```

```python
predictions #model predicted labels
```

```python
y_test #true lables
```

```python
from sklearn.metrics import accuracy_score
```

```python

accuracy_score(y_test, predictions)
```

```python
#simply pass ur training data to .score method
knn.score(X_test, y_test)
```

## findind the best fit model

```python
neighbors=range(1,31)
accuracies=[]
for i in neighbors:
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    acc=knn.score(X_test, y_test)
    accuracies.append(acc)
```

```python
accuracies
```

```python
import matplotlib.pyplot as plt
plt.plot(neighbors, accuracies)
```
