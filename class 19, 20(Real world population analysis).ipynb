```python
req=df[["Country name","Year","Population", "Population of children under the age of 15","Population aged 15 to 64 years"]]
```

```python
req["pop over 65"]=req["pop"]- (req["pop under 15"]+req["pop between 15-64"])
```

```python
# Define a function plot_population(frame, country) that takes the dataframe
# df and the name of one of its countries and returns a line chart of its total 
# population (an annual time series). For example, the this code produces the following figure:
```

```python
def plot_population(df, c):
    required=df[df["country"]==c]
    plt.plot(required["Year"], required["pop"])
    plt.title(c)
    plt.xlabel("Year")
    plt.ylabel("Population change")
    plt.grid(True)
    plt.yticks(ticks= [50000000, 100000000, 150000000, 200000000, 250000000, 300000000],labels=["50M", "100M", "150M", "200M", "250M", "300M"])
    plt.show()
```

```python
plot_population(req, "Pakistan")
```

```python
#Define a function plot_populations(frame, countries) that takes dataframe df and a list of countries 
#and returns a time-series plotting total population for those countries.
```

```python
def plot_lists(df, list_):
    for i in list_:
        required= df[df["country"]==i]
        plt.plot(required["Year"], required["pop"])
        plt.title("Population Comparison")
        plt.xlabel("Year")
        plt.ylabel("Population")
        plt.legend(list_)
        plt.grid(True)
        plt.yticks(ticks= [50000000, 100000000, 150000000, 200000000, 250000000, 300000000],labels=["50M", "100M", "150M", "200M", "250M", "300M"])
       
```

```python
plot_lists(req, ["Pakistan", "Albania", "Indonesia"])
```

```python
plot_lists(req, ["India", "China", "Pakistan"])
```

```python
plot_lists(req, ["South Korea", "China", "Japan"])
```

```python
def plot_percentages(df, c):
    filtered=df[df["country"]==c]
    bar_width=0.5
    years=filtered["Year"]
    fig, ax=plt.subplots(figsize=(10,7))
    ax.bar(years, filtered["under 15%"], bar_width, label="under 15%")
    ax.bar(years, filtered["15-64%"], bar_width, bottom=filtered["under 15%"], label="15-64 %")
    ax.bar(years, filtered["over 65%"], bar_width, bottom=filtered["under 15%"]+filtered["15-64%"], label="over 65%")
    ax.set_title("Population by age groups")
    ax.set_xlabel("Year")
    ax.set_ylabel("Population")
    ax.legend()
    ax.grid(True)
```

```python
plot_percentages(req, "Pakistan")
```

```python
plot_percentages(req,"India")
```
